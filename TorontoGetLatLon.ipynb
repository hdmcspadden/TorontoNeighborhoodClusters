## Get the Lat Lon for each Neighborhood

I am going to use the csv provided and merge the two DataFrames from two csvs:
1. I saved the web scraped postal coded neighborhoods to a csv
2. The csv provided with lat lons for each neighborhood/postal code

### First, I need to load the two DataFrames from csv files.

```python
# import libraries
import pandas as pd
```

```python
# load the csv of Toronto Neighborhoods created from the scraped Wikipedia webpage:
df_Neighborhoods = pd.read_csv('Toronto.csv')
print(df_Neighborhoods.shape)
df_Neighborhoods.head()
```

```python
# load the csv provided with lat lons for neighborhoods
df_LatLon = pd.read_csv('Geospatial_Coordinates.csv')
print(df_LatLon.shape)
df_LatLon.head()
```

Now I need to merge the two dataframes on the Postal Code column

```python
# rename the 'Postal Code' column to PostalCode in the df_LatLon DataFrame
df_LatLon.rename(columns={'Postal Code': 'PostalCode'}, inplace=True)

df_LatLon.head()
```

```python
print(df_LatLon.shape)
```

```python
# merge the DataFrames
df_Merged = df_Neighborhoods.merge(df_LatLon, how='left', on='PostalCode',)
df_Merged = df_Merged[['PostalCode','Borough','Neighborhood','Latitude','Longitude']]

# write the DataFrame to a csv to use in the next exercise
df_Merged.to_csv('TorontoNeighborhoodsLatLon.csv')

# print the shape, and print the dataframe
print(df_Merged.shape)
df_Merged
```
